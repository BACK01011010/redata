# 兰州天气数据探索

首先让我们详细查看数据的内容和结构。

```python
import pandas as pd
import numpy as np

# 读取数据
df = pd.read_excel('../数据集/lanzhou_weather_data.xlsx')

# 1. 显示数据基本信息
print("数据基本信息：")
print(df.info())

# 2. 显示所有列名
print("\n数据列名：")
print(df.columns.tolist())

# 3. 显示前10行数据
print("\n数据前10行：")
print(df.head(10))

# 4. 显示数据类型
print("\n数据类型：")
print(df.dtypes)

# 5. 显示每列的唯一值数量
print("\n每列唯一值数量：")
print(df.nunique())

# 6. 显示数值列的基本统计信息
print("\n数值列统计信息：")
print(df.describe())

# 7. 检查缺失值
print("\n缺失值统计：")
print(df.isnull().sum())

# 8. 显示每列的前几个唯一值
for column in df.columns:
    print(f"\n{column}的唯一值示例（前5个）：")
    print(df[column].unique()[:5])
```
