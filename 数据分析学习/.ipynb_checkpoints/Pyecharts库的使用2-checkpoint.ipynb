```python
# 整理数据：
data = [
    [234,240,249,210],
    [224,250,259,201],
    [211,230,247,219],
    [235,250,259,200],
]
```

```python
c = (
    Kline()
    .add_xaxis(['2024/12/1{}'.format(i+1) for i in range(4)])
    .add_yaxis('K线图',data)
    .set_global_opts(
        title_opts=opts.TitleOpts(title='K线图')
    )
)
c.render('可视化图表/K线图.html')
```

## 面积图

```python
import pyecharts.options as opts
from pyecharts.charts import Line
from pyecharts.faker import Faker
```

```python
c = (
    Line()
    .add_xaxis(Faker.choose())
    .add_yaxis('A',Faker.values(),is_smooth=True)
    .add_yaxis('B',Faker.values(),is_smooth=True)
    .add_yaxis('C',Faker.values(),is_smooth=True)
    # 展示面积部分
    .set_series_opts(
        areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
        label_opts=opts.LabelOpts(is_show=False)
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='面积图'),
        xaxis_opts=opts.AxisOpts(
            axistick_opts=opts.AxisTickOpts(is_align_with_label=True),
            is_scale=False,
            boundary_gap=False
        )
    )
)
c.render('可视化图表/面积图.html')
```

## 环形图

```python
c = (
    Pie()
    .add('动物园中动物比例',
         [list(z) for z in zip(Faker.choose(),Faker.values())],
         radius=['50%','80%']
        )
    .set_global_opts(
        title_opts=opts.TitleOpts(title='环形图')
    )
)
c.render('可视化图表/环形图.html')
```

```python
c = (
    Pie()
    .add('动物园当中动物比例',
         [list(z) for z in zip(Faker.choose(),Faker.values())],
         rosetype='radius',
         radius=['10%','80%'],
        )
)
c.render('可视化图表/玫瑰图.html')
```

```python
import pyecharts.options as opts
from pyecharts.charts import Pie,Timeline
from pyecharts.faker import Faker
```

```python
attr = Faker.choose()
t1 = Timeline()
for i in range(2020,2024):
    pie = (
        Pie()
        .add('商家',[list(z) for z in zip(attr,Faker.values())])
        .set_global_opts(title_opts=opts.TitleOpts(title='某商店{}年营业额'.format(i)))
    )
    t1.add(pie,'{}年'.format(i))
t1.render('可视化图表/时间轴组件-饼状图.html') 
```
