# DataFram

```python
# DataFram 是用于存储多行和多列的数据集合，
# 类似于excel的二维表
# 对于DataFram的操作，较多的就是 ‘增删改查’
```

```python
# 创建第一个数据框
```

```python
from pandas import DataFrame
```

```python
dic = {'name':['Andy','Jerry','Tom'],'tel':[1001,1002,1003]}
dic
```

```python
df1 = DataFrame(dic,index=['a','b','c'])
df1
```

```python
df2 = DataFrame(dic)
df2
```

```python
#创建一个数组，转换为数据框
```

```python
import numpy as np
arr = np.random.randint(10,15,size=(2,6))
arr
```

```python
df3 = DataFrame(arr,index=['a','b'])
df3
```

```python
# 练习：
# 把刚才的arr变量，转换为六行两列的数据，再次转换为数据框内容
```

```python
arr.shape = (6,2)
arr
```

```python
df4 = DataFrame(arr)
df4
```

```python
# 提取列的名称
df4.columns
```

```python
df4.columns = ['a','b']
df4
```

```python
# inplace=True  表示保存修改
df4.rename(columns={'a':'第一列'},inplace=True)
df4
```

```python
df4
```

```python
df5 = df4.reindex([1,2,5])
df5
```

```python
# 注：reindex()   我们可以理解为用此方法，可以提取某行的数据
```

```python
df5.index=['a','b','c']
df5
```

```python
df5.columns=['一','二']
df5
```

```python
# 练习：
# 通过numpy 生成课程表数据
# 一周五天，每天八节课
# 转为二维表，然后表头信息和序号进行修改
```

```python
import numpy as np
arr3 = np.random.randint(10,20,size=[8,5])
arr3
```

```python
dd1 = DataFrame(arr3)
dd1
```

```python
dd1.columns=['一','二','三','四','五']
dd1.index=['1','2','3','4','5','6','7','8']
dd1
```

```python
dd1.index = ["第{}节".format(i+1) for i in range(8)]
dd1
```

## 访问数据

```python
# 1.访问某一列数据
# 2.访问某一行数据
# 3.访问某一个值 (某一行列交叉位置)
# 4.访问某一块数据 (数据的多行多列交叉位置)
```

```python
df0.columns
```

```python
df0.index
```

```python
# 访问某一列数据
df0['姓名']
```

```python
# 只看名字，有时候不知道是男生还是女生
# 获取表格中，性别和姓名两列数据
df0[['姓名','性别']]
```

```python
# 2.访问某一行数据
df0.loc['0']
```

```python
# 3.访问某一个值 (某一行列交叉位置)
df0.at['0','体育']
```

```python
df0.at['5','数分']
```

```python
# 4.访问某一块数据 (数据的多行多列交叉位置)
df0.iloc[0:4,5:9]
```

```python
#头部几行数据，不写默认5行
df0.head(3) 
```

```python
#尾部几行数据，不写默认5行
df0.tail(3)
```

```python
# 增加行,与排序
df0.loc[20] = ['a','001','23080225','张三','男',1,1,1,1,1,1]
df0
```

```python
# 删除行和列
# 首先删除某一列数据
# inpalce=True 保存操作 df0.drop('Unnamed: 0',axis=1，inpalce=True) 
df0.drop('Unnamed: 0',axis=1) 
```

```python
# 首先删除某一行数据
df0.drop(20)
```
