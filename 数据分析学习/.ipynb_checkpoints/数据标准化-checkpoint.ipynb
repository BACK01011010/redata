# 数据标准化

常用的归一化方法

```python
# 数据的标准化（归一化）：
# 是数据分析和数据挖掘工作的一项基本工作，不同评价指标往往具有不同的
# 量纲和量纲单位，这样的情况下会影响到数据分析的结果
# 为了消除指标之间的量纲影响，需要进行数据标准化处理
```

```python
# min-max标准化
# x *= (x-min)/(max-min)
```

```python
import pandas as pd
df = pd.read_excel('data/data_df.xlsx',sheet_name='Sheet1')
df
```

```python
# 首先查看下军训的分数
print(df.军训.max(),df.军训.min(),sep='\n')
```

```python
# 提取列明
col = df.columns.tolist()
col
```

```python
col[7]
```

```python
# 然后获取：【'英语'，'体育，'军训'，'数分'，'高代'，'解几'】
col[5:]
```

```python
for i in col[5:]:
    print(i,df[i].max(),df[i].min(),sep='\t')
```

```python
# 对数据分列进行数据标准化，并且展示数据
# min-max标准化
# x*=(x-min)/(max-min)

# 首先对 ‘数据分析’ 列，进行数据标准化
# 再对'军训'列，进行数据标准化
```

```python
df.info()
```

```python
sf = df.数分.astype(int)
```

```python
scale = (sf-sf.min())/(sf.max()-sf.min())
scale
```

```python
jx = df.军训
```

```python
scalejx = (jx-jx.min())/(jx.max()-jx.min())
scalejx
```

```python
# 对所以课程进行数据标准化
for i in col[5:]:
    a = df[i].astype(float)
    df[i] = (a-a.min()) / (a.max()-a.min())
df
```
