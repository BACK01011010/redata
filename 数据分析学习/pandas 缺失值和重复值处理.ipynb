# 缺失值的处理

```python
# 由于工作中的各种原因，会纯在缺失值。
# 缺失值的存在可能会影响到后期 分析或者挖掘工作。
```

```python
# 删除法：当数据中的某个变量大部份值都是缺失值可以考虑删除、
# 填补法：对于连续变量，如果变量分布近似于正态分布，可以用均值替代缺失值
#         如果变量是有偏的情况，可以使用中位数去替代缺失值
#         如果变量是离散型数据，可以使用众数替换缺失值
# 插补法：可以使用决策树计算出来的预测值替换缺失值
```

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(1234)
d1 = np.random.randint(1,100,size=50)
d1 = d1.reshape(5,10)
df = pd.DataFrame(d1)
df
```

```python
df.columns = list('abcdefghij')
df
```

```python
df.a = [48,31,51,np.nan,47]
df.c = [39,np.nan,38,20,82]
df
```

删除缺失值
dropna()方法

```python
df.dropna()
```

填补缺失值 fillna()

```python
df.fillna(0)
```

```python
df.fillna('缺失值')
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
df.fillna({'a':0,'c':'缺考'})
```

```python
# 均值，
df_mean = df.mean()
df_mean
```

```python
# 中位数，
df_median = df.median()
df_median
```

```python
# 众数
df_mode = df.mode()
df_mode
```
