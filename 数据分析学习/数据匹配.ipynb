# 数据匹配

```python
# merge() 类似于excel中的vlookup()函数
```

```python
# 例子：
#     一张表中，有学号和姓名，
#     另一张表中，有学号和手机号。
#     中间排序可能不一样。
#     要求把两张表合并到一起去
```

```python
import pandas as pd
```

```python
st1 = pd.DataFrame({
    'id':[202403,202402,202401,202405],
    'name':['老张','老王','老李','老刘']
})
st2 = pd.DataFrame({
    'id':[202401,202402,202403,202404],
    'tel':[100,101,102,10003]
})
print(st1,st2,sep='\n\n')
```

```python
pd.merge(st1,st2,on='id',how='left')
```

```python
pd.merge(st1,st2,on='id',how='right')
```

```python
pd.merge(st1,st2,on='id',how='outer')
```
