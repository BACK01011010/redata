```python
from pyecharts.charts import Bar
from pyecharts import options as opts
# 默认自带数据
from pyecharts.faker import Faker
```

## 柱状图设计

```python
c = (
    Bar()
   .add_xaxis(['苹果','车厘子','香蕉','草莓','葡萄'])
   .add_yaxis('商店A水果销量',[115,150,120,130,160])
   .add_yaxis('商店B水果销量',[115,150,120,130,160])
   .set_global_opts(title_opts=opts.TitleOpts(title='商店水果销售情况表',subtitle='副标题'))
   .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),
        markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(y=120,name='销售预期')]),  # 此处添加逗号
        itemstyle_opts={
            "normal":{
                # "color":"#5470C6",
                "barBorderRadius": [5, 5, 0, 0]
            }
        }
    )
)
c.render('可视化图表/柱状图1.html')
```

日历图

```python
import datetime
import random
from pyecharts import options as opts
from pyecharts.charts import Calendar
```

```python
# 整理数据
begin = datetime.date(2025,1,1)
end = datetime.date(2025,12,31)
# 使用随机数表示每天的步数，展示在日历图当中
data = [[str(begin + datetime.timedelta(days=i)),
     random.randint(1000,25000),]
     for i in range((end-begin).days + 1)]
data
```

```python
c = (
    Calendar()
   .add('', data, calendar_opts=opts.CalendarOpts(range_=2025))
    # 加入热力图，优化数据展示
   .set_global_opts(
        title_opts=opts.TitleOpts(title='2025年每天步数情况'),
        visualmap_opts=opts.VisualMapOpts(
            max_=25000,
            min_=1000,
            orient='horizontal'
        )
    )
)
c.render('可视化图表/2025年日历图.html')
```
